#### Importando dados

```python
dados = pd.read_csv("produtos.csv")
display(dados)
```

#### Cadastrar operações

```python
# Passo 4 - Cadastrar as operacoes
for linha in dados.index:
    pyautogui.click(x=432, y=388) # clicar no 1º item
    
    

    time.sleep(3)
    codigo = dados.loc[linha,"codigo"]
    marca = dados.loc[linha,"marca"]
    tipo = dados.loc[linha,"tipo"]
    categoria = dados.loc[linha,"categoria"]
    preco_unitario = dados.loc[linha,"preco_unitario"]
    custo = dados.loc[linha,"custo"]

    # Preencher os campos
    pyautogui.write(str(dados.loc[linha,"codigo"]))
    time.sleep(2)
    pyautogui.press("tab")
    pyautogui.write(str(dados.loc[linha,"marca"]))
    time.sleep(2)
    pyautogui.press("tab")
    pyautogui.write(str(dados.loc[linha,"tipo"]))
    time.sleep(2)
    pyautogui.press("tab")
    pyautogui.write(str(dados.loc[linha,"categoria"]))
    time.sleep(2)
    pyautogui.press("tab")
    pyautogui.write(str(dados.loc[linha,"preco_unitario"]))
    time.sleep(2)
    pyautogui.press("tab")
    pyautogui.write(str(dados.loc[linha,"custo"]))
    time.sleep(2)
    pyautogui.press("tab")
    pyautogui.press("enter")
    time.sleep(2)
    pyautogui.click(x=450, y=275)
    time.sleep(3)
    

```
